```python
import json
from my_utils import flatten_dict
from connections import get_mongo_client
```

```python
client = get_mongo_client()
db = client['task']
collection = db['task']
```

```python
sample_doc = collection.find_one()
```

```python
flattened_fields = flatten_dict(sample_doc).keys()
```

```python
substring = "203449255"
```

```python
query = {
    "$or": [
        {field: {"$regex": substring, "$options": "i"}} for field in flattened_fields if field != "_id"
    ]
}
```

```python
results = collection.find(query).limit(2)
```

```python
output_file = "matching_documents.json"
with open(output_file, "w") as file:
    json.dump(list(results), file, default=str, indent=4)
```
