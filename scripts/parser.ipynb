```python
#################rubish so far###################





import pandas as pd

def flatten_json_field(json_data, target_field):

    def get_nested_value(data, keys):

        for key in keys:
            if isinstance(data, dict):
                data = data.get(key, None)
            elif isinstance(data, list):
                return data  # Return the list if encountered
            else:
                return None
        return data

    def expand_row(outer_data, nested_data, parent_key=""):

        flat_row = {}
        for key, value in outer_data.items():
            flat_row[key] = value  # Keep outer-level values
        if isinstance(nested_data, dict):
            for key, value in nested_data.items():
                flat_row[f"{parent_key}{key}"] = value
        return flat_row

    # Parse the target field path
    target_keys = target_field.split(".")

    # Prepare a list for storing rows
    rows = []

    # Iterate over documents in the JSON file
    for doc in json_data.get("output", []):
        # Extract the target field
        nested_data = get_nested_value(doc, target_keys)

        # Handle lists of dictionaries
        if isinstance(nested_data, list):
            for idx, item in enumerate(nested_data):
                outer_data = {key: value for key, value in doc.items() if key not in target_keys}  # Outer fields
                row = expand_row(outer_data, item, f"{target_field}[{idx}].")
                rows.append(row)

        # Handle single dictionary
        elif isinstance(nested_data, dict):
            outer_data = {key: value for key, value in doc.items() if key not in target_keys}  # Outer fields
            row = expand_row(outer_data, nested_data, f"{target_field}.")
            rows.append(row)

    # Convert rows to DataFrame
    df = pd.DataFrame(rows)
    return df
```
