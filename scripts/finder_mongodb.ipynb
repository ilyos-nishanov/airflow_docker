```python
import json
import pandas as pd
from time import time
from connections import get_mongo_client
from my_utils import insert_into_mssql, load_my_columns, \
                        get_numbers,map_dff_to_my_columns, max_number_find
```

```python
client = client = get_mongo_client()
db = client['task']
task_collection = db['task']
query = {
    'data.katm_077.return.data.credit_requests.credit_request': {'$exists': True}
    # ,'number': {'$eq': 7090166}
    # ,'number': {'$gt': max_num}
    ,'number': {'$gt': 5000000, '$lt': 5001000}
}
projection = {
    'number': 1,
    'data.katm_077.return.data.credit_requests.credit_request': 1
}
```

```python
docs = task_collection.find(query, projection)#.sort('number', -1)
```

```python
columns_file = 'katm_077_credit_reqs_fields.txt'
columns = load_my_columns(columns_file)
```

```python
for doc in docs:
    rows = []
    _id = str(doc.get('_id')) 
    number = doc.get('number')
    fields = doc.get('data', {}).get('katm_077', {}).get('return', {}).get('data', {}).get('credit_requrests', {}).get('credit_request', [])
    if isinstance(fields, dict):
        fields = [fields]
    elif fields is None:
        fields = []
    for field in fields:
        row = {'_id': _id, 'number': number, **field}
        rows.append(row)

    df = pd.DataFrame(rows)
    final_df = map_dff_to_my_columns(df, columns)
```
